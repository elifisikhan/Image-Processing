```python
import matplotlib.pyplot as plt
```

```python
im_1=plt.imread('kedii.jfif')
plt.imshow(im_1)
plt.show()
```

```python
im_1.ndim, im_1.shape
```

```python
m,n,p=im_1.shape
im_2=np.zeros((m,n,3),dtype=int)
m,n,im_2.shape
for m in range(im_1.shape[0]):
    for n in range(im_1.shape[1]):
        im_2[m,n,0]=im_1[m,n,0]-2
        im_2[m,n,1]=im_1[m,n,1]+50
        im_2[m,n,2]=im_1[m,n,2]-35
plt.imshow(im_2)
plt.show()
```

```python
def renkDegerleriniDegistir(im,s):
    im_1=im
    m,n,p=im_1.shape
    im_2=np.zeros((m,n,3),dtype=int)
    m,n,im_2.shape
    for m in range(im_1.shape[0]):
        for n in range(im_1.shape[1]):
            im_2[m,n,0]=im_1[m,n,0]-s
            im_2[m,n,1]=im_1[m,n,1]+s
            im_2[m,n,2]=im_1[m,n,2]-s
    return im
```

```python
def graySatirSutunKes(im):
    m,n,p=im.shape
    new_m=int(m/2)
    new_n=int(n/2)
    im_5=np.zeros((new_m,new_n),dtype=int)
    
    for m in range(new_m):
        for n in range(new_n):
            s0=(im[m*2,n*2,0]+im[m*2,n*2,1]+im[m*2,n*2,2])/3
            im_5[m,n]=int(s0)
    return im_5
```

```python
im_6=graySatirSutunKes(im_1)
plt.imshow(im_6,cmap='gray')
plt.show()
plt.imsave('kediGray.jpg',im_6,cmap='gray')
```

```python
def renkliSatirSutunKes(im):
    m,n,p=im.shape
    new_m=int(m/2)
    new_n=int(n/2)
    im_7=np.zeros((new_m,new_n,3),dtype=int)
    for m in range(new_m):
        for n in range(new_n):
            im_7[m,n,0]=int(im[m*2,n*2,0])
            im_7[m,n,1]=int(im[m*2,n*2,1])
            im_7[m,n,2]=int(im[m*2,n*2,2])
    return im_7
```
